```python
import requests
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
countries=pd.read_excel('UNSD — Methodology.xlsx', sheetname='Sheet1')
```

```python
countries1=countries.dropna(axis=0, subset=['ISO-alpha3 Code'])
```

```python
countries2=countries1[['Sub-region Name','ISO-alpha3 Code', 'Country or Area']]
countries2=countries2.set_index('ISO-alpha3 Code')
```

```python
countries2
```

```python
def GetClimateDataAllCountries(countries):
    dataframe=countries.copy()
    for country in dataframe.index:
        try:
            link="http://climatedataapi.worldbank.org/climateweb/rest/v1/country/cru/tas/year/"+country
            r=requests.get(link)
            country_data=pd.DataFrame(r.json())
            country_data=country_data.set_index('year').T
            country_data=country_data.rename({'data':country})
            dataframe=dataframe.combine_first(country_data)
        except (KeyError):
            print('KeyError for', country)
            dataframe=dataframe.drop(country)
            print('Removed', country, 'from dataframe.')
            pass
    return dataframe
```

```python
trade_data_export=trade_data_export.loc[(trade_data_export["Product categories"]=='All Products') & (trade_data_export["Indicator Type"]=='Export')]
  
```

```python
years = list(range(1995,2016))
countries=list(trade_data_export['Reporter'].unique())
```

```python
"""MultiIndex array is three-dimensional: year*country*country
""" 
#magically make MultiIndex slicing better
idx = pd.IndexSlice

#build MultiIndex array
rows_array = []
for year in years:
    for country in countries:
        rows_array.append([year,country])
        
multi_index = pd.MultiIndex.from_tuples(rows_array, names=['year', 'country'])
data = pd.DataFrame(columns=countries,index=multi_index)

```
