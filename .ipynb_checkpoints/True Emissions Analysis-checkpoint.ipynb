```python
import requests
import pandas as pd
from pathlib import Path
import numpy as np
```

```python
countries=pd.read_excel('UNSD — Methodology.xlsx', sheetname='Sheet1')
```

```python
countries1=countries.dropna(axis=0, subset=['ISO-alpha3 Code'])
```

```python
countries2=countries1[['Sub-region Name','ISO-alpha3 Code', 'Country or Area']]
countries2=countries2.set_index('ISO-alpha3 Code')
```

```python
countries2
```

```python
def GetClimateDataAllCountries(countries):
    dataframe=countries.copy()
    for country in dataframe.index:
        try:
            link="http://climatedataapi.worldbank.org/climateweb/rest/v1/country/cru/tas/year/"+country
            r=requests.get(link)
            country_data=pd.DataFrame(r.json())
            country_data=country_data.set_index('year').T
            country_data=country_data.rename({'data':country})
            dataframe=dataframe.combine_first(country_data)
        except (KeyError):
            print('KeyError for', country)
            dataframe=dataframe.drop(country)
            print('Removed', country, 'from dataframe.')
            pass
    return dataframe
```

```python
pathlist = Path('wits_en_trade_summary_allcountries_allyears').glob('**/*.CSV')
trade_data=pd.DataFrame()
for path in pathlist:
    # because path is object not string
    path_in_str = str(path)
    trade_data_country=pd.read_csv(path_in_str,encoding = "ISO-8859-1")
    trade_data=trade_data.append(trade_data_country)
    # print(path_in_str)
```

```python
trade_data_export = trade_data[trade_data.Partner != '...']
```

```python
trade_data_export=trade_data_export[trade_data_export['Reporter'].notnull()]
```

```python
trade_data_export['Reporter'].unique()
```

```python
countries_to_remove=['Belgium-Luxembourg', 'European Union', 'Other Asia, nes',' World']
```

```python
for country in countries_to_remove:
    trade_data_export = trade_data_export[trade_data_export.Reporter != country]
```

```python
len(trade_data_export['Reporter'].unique())
```

```python
trade_data_export
```

```python
years = list(range(1995,2016))
countries=list(trade_data_export['Reporter'].unique())
```

```python
"""MultiIndex array is three-dimensional: year*country*country
""" 
#magically make MultiIndex slicing better
idx = pd.IndexSlice

#build MultiIndex array
rows_array = []
for year in years:
    for country in countries:
        rows_array.append([year,country])
        
multi_index = pd.MultiIndex.from_tuples(rows_array, names=['year', 'country'])
data = pd.DataFrame(columns=countries,index=multi_index)

```

```python
data.loc[1995]
```

```python
for year in years:
    for export_country in countries:
        for import_country in countries:
            data.loc[year][export_country][import_country]=trade_data_export.loc[(trade_data_export["Reporter"]==export_country) & (trade_data_export["Partner"]==import_country)]
        
```
