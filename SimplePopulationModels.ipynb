## Run the model for each country

We want to run the core SD model for each country, with country specific parameterization. 

Thus, we formulate a function that based on each row parameterizes the model with the value from geodata, performs the simulation and finally returns the population over time for each country.

```python
def runner(row):
    sim_time=200
    params={'initial population':row['POP2005']}
    res=model.run(params=params, return_timestamps=np.linspace(0,sim_time,num=sim_time*2+1))
    return res['population']
```

```python
res=geo_data.apply(runner, axis=1)
```
